```python
import numpy as np
import heapq as hq
from copy import deepcopy
filename = "input_16"
#filename = "example_16.txt"
filename = "example_16_2.txt"
with open(filename) as f:
    input = [[i for i in line[1:-1]] for line in f.read().split("\n")[1:-1]]
```

```python
def find_shortest_path(input):

    xlen = len(input[0])
    ylen = len(input)
    directions = {(0,1), (0,-1), (1,0), (-1,0), (0,0)}
    score_map = {True: 1, False: 1001}
    direction_map = {0:(0,1), 1:(0,-1), 2:(1,0), 3:(-1,0)}
    grid = [input for _ in range(4)]

    c = 0 # use this counter to remove ambiguities in heap ordering

    # prepare input:
    i0, j0 = tuple(i[0] for i in np.where(np.array(input) == "S")) #starting
    iN, jN = tuple(i[0] for i in np.where(np.array(input) == "E")) #end
    for i in range(xlen):
        for j in range(ylen):
            for d in range(4):
                if input[i][j] == "#":
                    cost = float("inf")
                else:
                    cost = 1
                grid[d][i][j] = {
                    "cost" : cost,
                    "position": (i,j,d),
                    "distance": float("inf"),
                    "predecessor": None,
                }
    grid[0][i0][j0]["distance"] = 0
    heap = [(0, 0, grid[0][i0][j0])]
    seen = set()

    while heap:
        current_dist, _, current_node = hq.heappop(heap)
        current_position = current_node["position"]
        if current_position in seen: continue
        seen.add(current_position)
        current_dir = direction_map[current_position[2]]

        for y,x in directions:
            i, j = current_position[0] + y, current_position[1] + x
            for d in range(4):
                if i in range(ylen) and j in range(xlen) and [i,j] != current_node["predecessor"]:
                    node = grid[d][i][j]
                    dist = current_node["distance"] + node["cost"] * score_map[(y,x) == current_dir]
                    if current_dir != direction_map[d]:
                        dist += 1000
                    if dist < node["distance"]:
                        grid[d][i][j]["distance"] = dist
                        grid[d][i][j]["predecessor"] = current_position[:2]
                        hq.heappush(heap, (dist, (c:=c+1), grid[d][i][j]))
    print(grid)
    return grid
```
